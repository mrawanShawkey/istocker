```python
import pandas as pd
from pathlib import Path
import sys  
import numpy as np
ROOT_DIR = Path().resolve().parents[0]
sys.path.append(str(ROOT_DIR))
from config.paths import LSTM_RESULTS_FILE 
```

```python
df = pd.read_csv(LSTM_RESULTS_FILE)
```

```python
df.describe()
```

```python
print("\n=== AVERAGE PERFORMANCE ===")
print("Raw Sharpe :", df["raw_sharpe"].mean())
print("Z Sharpe   :", df["z_sharpe"].mean())
print("Rank Sharpe:", df["rank_sharpe"].mean())
```

```python
print("\n=== STABILITY (STD) ===")
print("Raw std :", df["raw_sharpe"].std())
print("Z std   :", df["z_sharpe"].std())
print("Rank std:", df["rank_sharpe"].std())
```

```python
print("\n=== WORST FOLD ===")
print("Raw worst :", df["raw_sharpe"].min())
print("Z worst   :", df["z_sharpe"].min())
print("Rank worst:", df["rank_sharpe"].min())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))

plt.plot(df["raw_sharpe"], label="Raw")
plt.plot(df["z_sharpe"], label="Z-score")
plt.plot(df["rank_sharpe"], label="Rank")

plt.axhline(0, linestyle="--")

plt.legend()
plt.title("Sharpe per Fold")
plt.show()  
```
